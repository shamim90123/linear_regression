```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
house_size = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)  # Reshaping for sklearn
house_price = np.array([200, 250, 300, 350, 400])

# Plot the data
plt.scatter(house_size, house_price, color='blue')
plt.title('House Size vs Price')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price (in $1000s)')
plt.show()

# Train the Linear Regression model
model = LinearRegression()
model.fit(house_size, house_price)

# Print the slope and intercept
print(f"Slope (m): {model.coef_[0]}")
print(f"Intercept (b): {model.intercept_}")

# Predict for a house of size 2800 sq ft
predicted_price = model.predict([[2800]])
print(f"Predicted price for a house with size 2800 sq ft: {predicted_price[0]} (in $1000s)")

# Plot the regression line
plt.scatter(house_size, house_price, color='blue')
plt.plot(house_size, model.predict(house_size), color='red')
plt.title('House Size vs Price (with Regression Line)')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price (in $1000s)')
plt.show()
```
